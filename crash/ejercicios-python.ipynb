## Condiciones

1. Escribe un programa que determine si un número es positivo, negativo o cero.

````python
n=int(input("Dame un número")) #con input se guarda como cadena de forma predeterminada y por eso se pone un int antes
if n<0:
    print("El número es negativo")
elif n>0:
    print("El número es positivo")
else:
    print("El número es cero")
````

2. Verifica si un número es par o impar.

````python
n=int(input("Dame un número")) #con input se guarda como cadena de forma predeterminada y por eso se pone un int antes
if n%2==0:
    print("El número es par")
else:
    print("El número es impar")
````

8. Crea un diccionario que almacene nombres de estudiantes como claves y sus notas como valores, luego calcula el promedio.

````python

nombre=input("Dame el nombre del estudiante")
nota=float(input("Dame su nota"))
dic3={nombre:nota}
suma=0
for i in range(4):
  nombre=input("Dame el nombre del estudiante")
  nota=float(input("Dame su nota"))
  dic3[nombre]=nota
for i in dic3.values():
  suma=suma+i
  print(suma)
notamedia=suma/len(dic3)
print(notamedia)


````
